```python
#what is the survival rate in each Pclass?
temp_data=train_data.loc[:,["Survived","Pclass"]].copy()
temp_data.groupby(["Pclass"]).mean()
```

```python
#what is the survival rate of each sex group?
temp_data=train_data.loc[:,["Survived","Sex"]].copy()
temp_data.groupby(["Sex"]).mean()
```

```python
#what is the survial rate of each age group (0-10,10-20,etc.) ?
temp_data = train_data.loc[:,["Survived","Age"]].copy()
#replace missing Age with 99999
temp_data["Age"].fillna(99999, inplace=True)

temp_data["AgeGroup"]=(temp_data["Age"]//10)*10

temp_data.drop(columns=["Age"],inplace=True)    

temp_data.groupby(["AgeGroup"]).mean()
```

```python
#what is the survival rate of each embarked location?
temp_data=train_data.loc[:,["Survived","Embarked"]].copy()
temp_data.groupby(["Embarked"]).mean()
```

```python
#what is the survival rate for each SibSp?
temp_data=train_data.loc[:,["Survived","SibSp"]].copy()
temp_data.groupby(["SibSp"]).mean()
```

```python
#what is the survival rate for each Parch?
temp_data=train_data.loc[:,["Survived","Parch"]].copy()
temp_data.groupby(["Parch"]).mean()
```
